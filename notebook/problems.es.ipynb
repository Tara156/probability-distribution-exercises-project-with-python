# Problemas de distribución de probabilidad

## Distribución binomial

En teoría de la probabilidad y estadística, la distribución binomial es la distribución de probabilidad discreta que da sólo dos resultados posibles en un experimento; por ejemplo, al lanzar una moneda, puede salir cara o cruz.

```python
from scipy.stats import binom

# Definir todas las funciones de probabilidad 
# relacionadas con esta distribución.

def dbinom(x, size, prob = 0.5):
    """
    Calcula la estimación puntual de la distribución binomial.
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calcula el acumulado de la distribución binomial.
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calcula la función cuantil a partir de la distribución binomial.
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Genera variables aleatorias a partir de la distribución binomial.
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Distribución binomial(10, 0.2), dónde n = 10 es el número de ensayos y p = 0.2 es la probabilidad de éxito en un solo ensayo
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
print("Normal: P(X=0) ≈", dnorm(0))                 # ~0.3989
print("Normal: P(X≤1.96) =", pnorm(1.96))           # ~0.975
print("Normal: Cuantil 0.95 =", qnorm(0.95))        # ~1.645
print("Normal: Muestras:", rnorm(5))
```

```python
from scipy.stats import uniform

def dunif(x, min_val=0, max_val=1):
    """
    Densidad de probabilidad de la distribución uniforme continua.
    """
    return uniform.pdf(x, loc=min_val, scale=max_val - min_val)

def punif(x, min_val=0, max_val=1):
    """
    Probabilidad acumulada de la distribución uniforme continua.
    """
    return uniform.cdf(x, loc=min_val, scale=max_val - min_val)

def qunif(p, min_val=0, max_val=1):
    """
    Cuantil de la distribución uniforme continua.
    """
    return uniform.ppf(p, loc=min_val, scale=max_val - min_val)

def runif(n, min_val=0, max_val=1):
    """
    Genera variables aleatorias uniformes continuas.
    """
    return uniform.rvs(loc=min_val, scale=max_val - min_val, size=n)
```

```python
# Semilla
np.random.seed(42)

# Parámetros de la distribución uniforme
min_val = 2
max_val = 10

# Generar datos aleatorios desde la distribución uniforme continua
dataa = runif(2000, min_val=min_val, max_val=max_val)

# Evaluar las funciones
pmfa = dunif(dataa, min_val=min_val, max_val=max_val)
cdfa = punif(dataa, min_val=min_val, max_val=max_val)
ppfa = qunif(cdfa, min_val=min_val, max_val=max_val)  # inversa de la acumulada

# Crear subgráficos
fig, axis = plt.subplots(3, 1, figsize=(6, 9))

axis[0].scatter(dataa, pmfa, color="red", alpha=0.5)
axis[0].set_title("PDF - Densidad de probabilidad (dunif)")

axis[1].scatter(dataa, cdfa, color="blue", alpha=0.5)
axis[1].set_title("CDF - Probabilidad acumulada (punif)")

axis[2].scatter(dataa, ppfa, color="green", alpha=0.5)
axis[2].set_title("Función cuantil (qunif)")

plt.tight_layout()
plt.show()
```
